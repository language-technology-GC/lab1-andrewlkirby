```python
#Edited 10/12/20
with gzip.open('news.2007.en.shuffled.deduped.gz', 'rt') as fin, gzip.open('tokens.gz', 'wt') as fout:
    for line in fin:
        tokens = word_tokenize(line)
        print(' '.join(tokens), file=fout)
print("Done!")
```

```python
#Edited 10/12/2020
#Putting first 1000 lines to my sample file:
import itertools

with gzip.open('tokens.gz', 'rt') as fin, open('tokensample.txt', 'w') as fout:
    for line in itertools.islice(fin, 1000):
        print(line, file=fout)
print("Done!")
```

```python
dfppmi = pd.read_csv(r"results.tsv", sep='\t', names=["w1", "w2", "PPMI score"])
dfppmi
```

```python
dfw2v = pd.read_csv(r'resultsw2v.tsv', sep='\t', names=["w1", "w2", "W2V score"])
dfw2v
```

```python
df = pd.read_csv(r"ws353.tsv", sep='\t', names=["w1", "w2", "score"])
```

```python
df2 = pd.merge(df, dfw2v)
df2
```

```python
import numpy
from scipy import stats
scorearr = df2['score'].to_numpy()
w2varr = df2['W2V score'].to_numpy()

def spear(x, y):
    return stats.spearmanr(x, y)

print(spear(scorearr, w2varr))
```

```python
df3 = dfppmi.merge(df, how='inner', on = ['w1', 'w2'])
df3
```

```python
ppmiarr = df3['PPMI score'].to_numpy
scorearr1 = df3['score'].to_numpy
print(spear(scorearr1, ppmiarr))
```
